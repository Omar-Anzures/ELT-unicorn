```python
#librerias que se van a utilizar 
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#dataset limpio
data = pd.read_csv('archivo.csv')
datos = pd.DataFrame(data)
```

```python
y = [0,50,100,150,200,250,300,400]
sns.set(style="whitegrid")
a = datos.head()

valores = sns.barplot(a.Industry,a.Valuation)
valores.set_xticklabels(a.Industry,rotation=80)
valores.set_yticklabels(y)
```

```python
import pandas as pd 
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('archivo.csv')
datos = pd.DataFrame(data)
```

```python
sns.set_style("whitegrid")
color = ['#0026e9']
sns.set(rc = {'figure.figsize':(12,8)})

conjunto = datos.groupby('Country')['Valuation'].sum()
d = conjunto.sort_values(ascending = False).head()    

ax = sns.barplot(d.values, d.index, palette = color)

for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(0)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f} mil'.format(p.get_width()/1000)
                ax.text(_x, _y,value, ha="left")
              
ax.set_title('Paises con mayor capital recibido')
ax.set_ylabel('Country',labelpad = 25)
ax.set_xlabel("Valuation",labelpad = 25)

#Eliminar los cuatro bordes:
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
```

```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('archivo.csv')
datos = pd.DataFrame(data)
```

```python
datos.Industry.replace(to_replace='Artificial intelligence', value='Artificial',inplace=True)
datos.Industry.replace(to_replace='Internet software & services', value='Internet',inplace=True)
datos.Industry.replace(to_replace='E-commerce & direct-to-consumer', value='E-commerce',inplace=True)
```

```python
sns.set_style("whitegrid")
color = ['#9b51e0']

conjunto = datos.groupby('Country')['Valuation'].sum()
d = conjunto.sort_values(ascending = False).head()    

ax = sns.barplot(d.values, d.index, palette = color,ci=None)

for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(0)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f} mil'.format(p.get_width())
                ax.text(_x, _y,value, ha="left")
              
ax.set_title('Industrias con mayor capital recibido')
ax.set_ylabel('Valuation',labelpad = 25)
ax.set_xlabel("Industry",labelpad = 25)

#Eliminar los cuatro bordes:
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
```
